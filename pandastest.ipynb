```python
import pandas as pd
import os
import datetime
import time
```

```python
cdi = os.getcwd()
cdir = cdi + r'\hist_amd.csv'
```

```python
df = pd.read_csv(cdir)
```

```python
df
```

```python
data = df.datetime
```

```python
d = [datetime.datetime.fromtimestamp(i/1000) for i in data]
```

```python
dt = pd.DataFrame(d,columns=['datetime'])
dt
```

```python
del df['datetime']
```

```python
amd = pd.concat([df,dt],axis=1)
```
